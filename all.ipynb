Rename

```python
import os

folder_path = './image'
list_categories = os.listdir(path=folder_path)
list_categories
```

```python
for category in list_categories:
    list_images = os.listdir(folder_path + '/' + category)

    for id, image in enumerate(list_images):
        os.rename(folder_path + '/' + category + '/' + image, folder_path + '/' + category + '/' + category + '_' + str(id + 1) + os.path.splitext(image)[-1])
```

Xử lý all

```python
from pic_to_square import covert_to_pixel_style
import cv2 as cv
output_folder = './all_output_low/'

for category in list_categories:
    list_images = os.listdir(folder_path + '/' + category)

    for id, image in enumerate(list_images):
        output_name = os.path.splitext(image)[0]
        print(output_name)
        img = cv.imread(folder_path + '/' + category + '/' + image)
        for color_num in range(20, 50, 10):
            for new_width in range(40, 90, 10):
                covert_to_pixel_style(img, color_num, new_width, output_folder + output_name)
```
